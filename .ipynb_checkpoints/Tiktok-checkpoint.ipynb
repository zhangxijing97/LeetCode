Function enqueue(call, queue, rear, maxSize):

Which of the following pseudocode correctly builds the Huffman tree?<br>
Pick ONE option.<be>
Option 1:
```
Function buildHuffmanTree(frequencies):
   Create a priority queue Q
   For each char in frequencies:
      Insert char into Q with frequency as priority
   While Q.size == 1:
      left = Q.extractMin()
      right = Q.extractMin()
      newNode = Merge left and right
      Insert newNode into Q
   Return Q.extractMin()
```
Option 2:
```
Function buildHuffmanTree(frequencies):

      Insert char into Q with frequency as priority

      left = Q.extractMin()
      right = Q.extractMin()
      newNode = Merge left and right
      Insert newNode into frequencies

```
Function buildHuffmanTree(frequencies):

Answer: Option 3: Correctly implements the Huffman tree building process. The while Q.size > 1 ensures nodes are merged until only one node (the root) remains in the queue. The Insert newNode into Q step ensures the priority queue Q always maintains the correct order based on frequencies.<br>

4. Object Lifecycle and Resource Management in a Banking System<br>

In a banking application, you are tasked with implementing a BankAccount class that manages customer accounts, including account opening and closing. The BankAccount class must handle external resources like file logging for transactions. Which of the following OOP design ensures that resources are properly cleaned up after use, even when exceptions occur?<br>

Pick ONE option:<br>

- Use of constructors to open resources and System.exit() in final blocks to close them.<br>
- Use of a destructor (__del__() in Python or finalize() in Java) to close resources when the object is garbage collected.<br>
- Implementing the AutoCloseable (Java) or __enter__ / __exit__ (Python) pattern for resource management.<br>
- Relying on the JVM or Python interpreterâ€™s garbage collector to automatically close all resources.<be>

Answer: 3. Implementing the AutoCloseable (Java) or __enter__ / __exit__ (Python) pattern for resource management.<br>
- System.exit():<br>

| **Aspect**               | **`System.exit()`**                                                                                                         | **`finalize()`**                                                                                              |
|---------------------------|-----------------------------------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------------------------|
| **Purpose**               | To **immediately terminate** the JVM and the program.                                                                        | To perform **cleanup tasks** for an object before it is garbage collected.                                   |
| **Scope**                 | Affects the **entire program** (terminates all threads and processes running in the JVM).                                     | Affects only a **single object** that is about to be garbage collected.                                       |
| **Trigger**               | Explicitly called in the code.                                                                                               | Implicitly invoked by the JVM **during garbage collection**, if implemented in the object class.             |
| **Determinism**           | **Deterministic**: Executes immediately when invoked.                                                                        | **Non-deterministic**: Called at the JVM's discretion, depending on when garbage collection occurs.           |
| **Impact on JVM**         | Terminates the JVM, stopping all ongoing processes and threads.                                                              | Does not terminate the JVM; operates within the JVM's lifecycle.                                              |
| **Use Case**              | Used to terminate the program during critical failures or when a specific exit code needs to be returned.                    | Used as a fallback for releasing unmanaged resources (e.g., file handles, sockets) if not explicitly closed. |
| **Resource Management**   | Does not handle resource cleanup. **Developers must manually clean up resources** before calling `System.exit()`.            | Used to clean up resources held by an object, but is unreliable for critical resource management.             |
| **Performance Impact**    | Minimal, as it directly halts execution and stops the JVM.                                                                   | Adds performance overhead, as objects with `finalize()` are queued for finalization before garbage collection. |

System.exit():
```
public static void main(String[] args) {
    System.out.println("Critical error: Exiting program...");
    System.exit(1); // Exit with error code
}
```
finalize():
```
class ResourceHandler {
    @Override
    protected void finalize() throws Throwable {
        System.out.println("Finalizing object...");
        // Perform cleanup here
        super.finalize();
    }
}

public static void main(String[] args) {
    ResourceHandler handler = new ResourceHandler();
    handler = null; // Make object eligible for garbage collection
    System.gc(); // Request garbage collection
}

```


| **Aspect**           | **`finalize()`**                                                            | **`AutoCloseable`**                                                    |
|-----------------------|-----------------------------------------------------------------------------|------------------------------------------------------------------------|
| **Purpose**           | Fallback cleanup when an object is garbage collected.                      | Deterministic cleanup of resources when used in a `try-with-resources` block. |
| **Trigger**           | Called by the JVM during garbage collection.                               | Explicitly called using `close()` at the end of the `try` block.       |
| **Determinism**       | **Non-deterministic**: No guarantee when or if it will run.                | **Deterministic**: Runs immediately when the block ends.              |
| **Performance**       | Adds overhead to garbage collection.                                       | Efficient and lightweight.                                             |
| **Recommended Use**   | Rarely used; discouraged in modern Java.                                   | Preferred for managing resources (e.g., files, streams, connections).  |
| **Example**           | `@Override protected void finalize() { ... }`                             | `class MyResource implements AutoCloseable { public void close() { ... } }` |


AutoCloseable:
```
class ResourceWithAutoCloseable implements AutoCloseable {
    @Override
    public void close() {
        System.out.println("Cleaning up in close()");
    }
}

public class AutoCloseableExample {
    public static void main(String[] args) {
        try (ResourceWithAutoCloseable resource = new ResourceWithAutoCloseable()) {
            System.out.println("Using the resource");
        } // Automatically calls close()
    }
}
```

Option 4: Relying on the JVM or Python interpreterâ€™s garbage collector to automatically close all resources.

Why itâ€™s incorrect:
Garbage collectors do not manage external resources like file handles or database connections.
Leaving resource cleanup to garbage collection can result in resource leaks and inconsistent behavior.

5. You are developing a class Order for an e-commerce platform. This class handles customer orders, including sensitive information like payment details. Which of the following best demonstrates proper encapsulation to protect sensitive data?<br>

- Make all attributes of the Order class public so that external classes can modify them directly.
- Use protected attributes and provide public getter and setter methods for all attributes.
- Declare sensitive attributes like paymentDetails as private and provide only necessary getter methods, without setters.
- Define all methods as private to ensure no other class can access the functionality.

Answer: Option 3: Declare sensitive attributes like paymentDetails as private and provide only necessary getter methods, without setters.<br>

````python
def countCreatorCommunities(related):
    def dfs(node, visited, n):
        # Mark the current node as visited
        visited[node] = True
        for neighbor in range(n):
            if related[node][neighbor] == '1' and not visited[neighbor]:
                dfs(neighbor, visited, n)
    
    n = len(related)  # Number of creators
    visited = [False] * n  # To track visited nodes
    communities = 0  # Count of distinct communities

    for i in range(n):
        if not visited[i]:
            # If a node is not visited, start a new DFS
            dfs(i, visited, n)
            communities += 1  # Increment community count
    return communities

# Example Input
related = [
    "1100",
    "1110",
    "0110",
    "0001"
]

# Function Call
print(countCreatorCommunities(related))  # Output: 2
````
