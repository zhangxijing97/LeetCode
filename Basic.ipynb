List is a collection which is ordered and changeable. Allows duplicate members.

Set is a collection which is unordered, unchangeable*, and unindexed. No duplicate members.

Dictionary is a collection which is ordered** and changeable. No duplicate members.

Tuple is a collection which is ordered and unchangeable. Allows duplicate members.

#### List  Methods and functions

| Method |   Action   |
|:---|:---|
|   len(x) | Returns the number of items in x   |
|   x.index(i) | Returns the index of items in x   |
|   min(x) | Returns the minimum value in x     |
|   max(x) | Returns the maximum of items in x   |
|   x.append(x1) | will append x1 to the end of x     |
|   x.extend(x1) | will extend the list x by appending all the elements i list x1, x1 has to be a list  |
|   x.insert(i,x1) | will insert item x1 at index i    |
|   x.count(i) | Returns frequency of occurence of i  |
|   x.remove(i)| Will remove the first i from x  will raise ValueError ifnthere is no i      |
|   x.pop([i]) | will remove and return the item at index i.  |
|   del x[a:b] | This method deletes all the elements in range starting from index ‘a’ till ‘b’ mentioned in arguments.    |
|   x.sort() | Will sort the items in A in-place   |
|   x.reverse() | Reverse the order of the list in place   |
|   all(list1) | True if all items in the list have a True value  |
|   any(list1) | True if one item in the list have a True value  |

```python
# List
# Creating a sample list
my_list = [5, 3, 8, 6, 2, 8, 9, 1]

# len(x)
print("Length of my_list:", len(my_list))  # Output: 8

# x.index(i)
print("Index of first occurrence of 8:", my_list.index(8))  # Output: 2

# min(x)
print("Minimum value in my_list:", min(my_list))  # Output: 1

# max(x)
print("Maximum value in my_list:", max(my_list))  # Output: 9

# x.append(x1)
my_list.append(4)
print("List after appending 4:", my_list)  # Output: [5, 3, 8, 6, 2, 8, 9, 1, 4]

# x.extend(x1)
my_list.extend([7, 10])
print("List after extending with [7, 10]:", my_list)  # Output: [5, 3, 8, 6, 2, 8, 9, 1, 4, 7, 10]

# x.insert(i, x1)
my_list.insert(2, 15)
print("List after inserting 15 at index 2:", my_list)  # Output: [5, 3, 15, 8, 6, 2, 8, 9, 1, 4, 7, 10]

# x.count(i)
print("Frequency of 8 in my_list:", my_list.count(8))  # Output: 2

# x.remove(i)
my_list.remove(8)
print("List after removing the first occurrence of 8:", my_list)  # Output: [5, 3, 15, 6, 2, 8, 9, 1, 4, 7, 10]

# x.pop([i])
popped_item = my_list.pop(3)
print("Item popped at index 3:", popped_item)  # Output: 6
print("List after popping item at index 3:", my_list)  # Output: [5, 3, 15, 2, 8, 9, 1, 4, 7, 10]

# del x[a:b]
del my_list[1:4]
print("List after deleting elements from index 1 to 4:", my_list)  # Output: [5, 8, 9, 1, 4, 7, 10]

# x.sort()
my_list.sort()
print("List after sorting:", my_list)  # Output: [1, 4, 5, 7, 8, 9, 10]

# x.reverse()
my_list.reverse()
print("List after reversing:", my_list)  # Output: [10, 9, 8, 7, 5, 4, 1]

# all(list1)
all_true = [True, True, True]
print("All items are True:", all(all_true))  # Output: True

some_false = [True, False, True]
print("Not all items are True:", all(some_false))  # Output: False

# any(list1)
print("At least one item is True:", any(some_false))  # Output: True

all_false = [False, False, False]
print("No item is True:", any(all_false))  # Output: False
```

### Set Methods and Functions

| Method | Action |
|:---|:---|
| `len(s)` | Returns the number of items in set `s` |
| `s.add(x)` | Adds item `x` to set `s` |
| `s.update([x1, x2, ...])` | Adds multiple items to set `s` |
| `s.remove(x)` | Removes item `x` from set `s`; raises KeyError if `x` is not found |
| `s.discard(x)` | Removes item `x` from set `s` if it is present |
| `s.pop()` | Removes and returns an arbitrary item from set `s`; raises KeyError if set is empty |
| `s.clear()` | Removes all items from set `s` |
| `s.copy()` | Returns a shallow copy of set `s` |
| `s.union(t)` | Returns a new set with items from both sets `s` and `t` |
| `s.intersection(t)` | Returns a new set with items common to sets `s` and `t` |
| `s.difference(t)` | Returns a new set with items in set `s` but not in set `t` |
| `s.symmetric_difference(t)` | Returns a new set with items in either set `s` or set `t` but not both |
| `s.issubset(t)` | Returns `True` if set `s` is a subset of set `t` |
| `s.issuperset(t)` | Returns `True` if set `s` is a superset of set `t` |
| `s.isdisjoint(t)` | Returns `True` if set `s` has no items in common with set `t` |

```python
# Creating a sample set
my_set = {1, 2, 3, 4, 5}

# len(s)
print("Length of my_set:", len(my_set))  # Output: 5

# s.add(x)
my_set.add(6)
print("Set after adding 6:", my_set)  # Output: {1, 2, 3, 4, 5, 6}

# s.update([x1, x2, ...])
my_set.update([7, 8, 9])
print("Set after adding multiple items [7, 8, 9]:", my_set)  # Output: {1, 2, 3, 4, 5, 6, 7, 8, 9}

# s.remove(x)
my_set.remove(2)
print("Set after removing 2:", my_set)  # Output: {1, 3, 4, 5, 6, 7, 8, 9}

# s.discard(x)
my_set.discard(3)
print("Set after discarding 3:", my_set)  # Output: {1, 4, 5, 6, 7, 8, 9}
# Discarding a non-existent item does not raise an error
my_set.discard(10)
print("Set after discarding non-existent item 10:", my_set)  # Output: {1, 4, 5, 6, 7, 8, 9}

# s.pop()
popped_item = my_set.pop()
print("Popped item:", popped_item)  # Output: 1 (or another arbitrary item)
print("Set after popping an item:", my_set)  # Output: {4, 5, 6, 7, 8, 9}

# s.clear()
my_set.clear()
print("Set after clearing all items:", my_set)  # Output: set()

# Copying and union/intersection/difference operations
my_set = {1, 2, 3}
another_set = {3, 4, 5}

# s.copy()
my_set_copy = my_set.copy()
print("Copy of my_set:", my_set_copy)  # Output: {1, 2, 3}

# s.union(t)
union_set = my_set.union(another_set)
print("Union of my_set and another_set:", union_set)  # Output: {1, 2, 3, 4, 5}

# s.intersection(t)
intersection_set = my_set.intersection(another_set)
print("Intersection of my_set and another_set:", intersection_set)  # Output: {3}

# s.difference(t)
difference_set = my_set.difference(another_set)
print("Difference of my_set and another_set:", difference_set)  # Output: {1, 2}

# s.symmetric_difference(t)
symmetric_difference_set = my_set.symmetric_difference(another_set)
print("Symmetric difference of my_set and another_set:", symmetric_difference_set)  # Output: {1, 2, 4, 5}

# s.issubset(t)
print("Is my_set a subset of another_set?:", my_set.issubset(another_set))  # Output: False

# s.issuperset(t)
print("Is my_set a superset of another_set?:", my_set.issuperset(another_set))  # Output: False

# s.isdisjoint(t)
print("Are my_set and another_set disjoint?:", my_set.isdisjoint(another_set))  # Output: False
```
